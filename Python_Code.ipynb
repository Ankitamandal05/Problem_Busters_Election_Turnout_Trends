```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: Voter turnout percentages by year and country
data = {
    'Year': [2000, 2005, 2010, 2015, 2020],
    'Country_A': [70, 68, 65, 67, 66],
    'Country_B': [60, 62, 61, 59, 58],
    'Country_C': [75, 74, 76, 77, 78]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
for country in df.columns[1:]:
    plt.plot(df['Year'], df[country], marker='o', label=country)

plt.title('Voter Turnout Trends Over Decades')
plt.xlabel('Year')
plt.ylabel('Turnout Percentage')
plt.legend()
plt.grid(True)
plt.savefig('voter_turnout_trends.png')
plt.show()
```
